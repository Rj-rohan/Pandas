```python
# Deleting Column
data.drop(['salary'],axis=1) # axis =1 for column and axis = 0 for row
```

```python
# Deleting ROW
data.drop([2],axis=0) # axis =1 for column and axis = 0 for row
```

```python
# Adddition
data['Addition_salary'] = data['salary'] + data['salary_in_usd']  # This will add "Addition_salary" column in dataset at last with addition
data
```

```python
# Subtraction
data['Subtraction_salary'] = data['salary'] - data['salary_in_usd']  # This will add "Addition_salary" column in dataset at last with addition
data   # Perform same for all arithmatic operations
```

```python
# MAX 
max_salary = data['salary'].max()
print("The maximum salary in salary column of given dataset is: ",max_salary)
```

```python
# MIN
min_salary = data['salary'].min()
print("The maximum salary in salary column of given dataset is: ",min_salary)
```

```python
# Mean
mean_salary = data['salary'].mean()
print("The maximum salary in salary column of given dataset is: ",mean_salary)
```

```python
# Median
median_salary = data['salary'].median()
print("The maximum salary in salary column of given dataset is: ",median_salary)
```
